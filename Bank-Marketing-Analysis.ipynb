##### Data Quality Assurance 

```python
#Checking unique values
df['age'].unique()
#Converting each value into Integer data type
df['age']=df['age'].astype(int)
#Checking counts and the data types of the values
df['age'].value_counts()

#Checking unique values
df['job'].unique()
# Repalced blue-collar with 'labor'
df['job'] =df['job'].replace( 'blue-collar','labor')
#Removing whiteshapes
df['job'] = df['job'].str.strip()
#Checking counts and the data types of the values
df['job'].value_counts()

#Replaced Unknown Value with possbile value 
#If the age is more than 60, Unknown job titles repalce with retired
df.loc[(df['age']>60) & (df['job']=='unknown'), 'job'] = 'retired'
#Replace education unknown value with possible value
df.loc[(df['education']=='unknown') & (df['job']=='management'), 'education'] = 'university.degree'
df.loc[(df['education']=='unknown') & (df['job']=='services'), 'education'] = 'high.school'
df.loc[(df['education']=='unknown') & (df['job']=='housemaid'), 'education'] = 'basic.4y'
#Replace job unknown value with possible value
df.loc[(df['job'] == 'unknown') & (df['education']=='basic.4y'), 'job'] = 'labor'
df.loc[(df['job'] == 'unknown') & (df['education']=='basic.6y'), 'job'] = 'labor'
df.loc[(df['job'] == 'unknown') & (df['education']=='basic.9y'), 'job'] = 'labor'
df.loc[(df['job']==  'unknown') & (df['education']=='professional.course'), 'job'] = 'technician'


# Remove all unknown value in job column 
index_names1 = df[ (df['job'] == 'unknown')].index 
df.drop(index_names1, inplace = True)


#Checking unique values
df['education'].unique()
#Removing whiteshapes
df['education'] = df['education'].str.strip()
#Checking counts and the data types of the values
df['education'].value_counts()
#Remove all unknown value in education column 
index_names2 = df[ (df['education'] == 'unknown')].index 
df.drop(index_names2, inplace = True)

#Grouping education with the similar qualification 
df['education'] = df['education'].replace(['basic.4y','basic.6y',
                                         'basic.9y'],'Primary')

df['education'] = df['education'].replace(['high.school'],'Secondary')
df['education'] = df['education'].replace(['professional.course'],'College')


df['education'] = df['education'].replace(['university.degree'],'Tertiary ')


#Remove all unknown value in marital column
df['marital'].unique()
#Removing whiteshapes
df['marital'] = df['marital'].str.strip()
#Checking counts and the data types of the values
df['marital'].value_counts()

#Grouping Marital Status with the similar types present
df['marital'] = df['marital'].replace(['divorced'],'single')
#Remove all unknown value in marital column
index_names3 = df[ (df['marital'] == 'unknown')].index 
df.drop(index_names3, inplace = True)

#Checking unique values
df['default'].unique()
#Removing whiteshapes
df['default'] = df['default'].str.strip()
#Checking counts and the data types of the values
df['default'].value_counts()
#Remove all unknown value in default column
index_names4 = df[ (df['default'] == 'unknown')].index 
df.drop(index_names4, inplace = True)


#Checking unique values
df['housing'].unique()
#Removing whiteshapes
df['housing'] = df['housing'].str.strip()
#Checking counts and the data types of the values
df['housing'].value_counts()


#Checking unique values
df['loan'].unique()
#Removing whiteshapes
df['loan'] = df['loan'].str.strip()
#Checking counts and the data types of the values
df['loan'].value_counts()
#Remove all unknown value in loan column
index_names5 = df[ (df['loan'] == 'unknown')].index 
df.drop(index_names5, inplace = True)

#Checking unique values
df['contact'].unique()
#Removing whiteshapes
df['contact'] = df['contact'].str.strip()
#Checking counts and the data types of the values
df['contact'].value_counts()


#Checking unique values
df['day_of_week'].unique()
#Removing whiteshapes
df['day_of_week'] = df['day_of_week'].str.strip()
#Checking counts and the data types of the values
df['day_of_week'].value_counts()

#Checking unique values
df['month'].unique()
#Removing whiteshapes
df['month'] = df['month'].str.strip()
#Checking counts and the data types of the values
df['month'].value_counts()


#Rename the column
df.rename(columns={"emp.var.rate":"emp_var_rate",
                   "cons.price.idx":"cons_price_idx",
                   "cons.conf.idx":"cons_conf_idx",
                   "nr.employed":"nr_employed",
                 "y":"deposit" }
          ,inplace=True)


```

```python
def cross_tab(df,f1,f2):
    jobs=list(df[f1].unique())
    edu=list(df[f2].unique())
    dataframes=[]
    for e in edu:
        dfe=df[df[f2]==e]
        dfejob=dfe.groupby(f1).count()[f2]
        dataframes.append(dfejob)
    xx=pd.concat(dataframes,axis=1)
    xx.columns=edu
    xx=xx.fillna(0)
    return xx
```

```python
cross_tab(df,'job','education')
```

```python

class UnderstandingData:
    
    def __init__(self, raw_df):
        self.raw_df = raw_df
        self.raw_df_grouped = raw_df.groupby("deposit")
        self.class_name_no = "no"
        
        self.class_name_yes = "yes"
        self.raw_df_grouped_no = self.raw_df_grouped.get_group(self.class_name_no)
        self.raw_df_grouped_yes = self.raw_df_grouped.get_group(self.class_name_yes)

    def plot_histogram_continuous(self, feature_name, bin_size):
        plt.figure()
        plt.hist(self.raw_df_grouped_no[feature_name], bins=bin_size, label=self.class_name_no)
        plt.hist(self.raw_df_grouped_yes[feature_name], bins=bin_size, label=self.class_name_yes)
        plt.legend()
        plt.title("Feature Histogram - "+feature_name)
        plt.xlabel("Feature values")
        plt.ylabel("Count")

    def plot_histogram_categorical(self, feature_name):
        feature_df = pd.DataFrame()
        feature_df["no"] = self.raw_df_grouped_no[feature_name].value_counts()
        feature_df["yes"] = self.raw_df_grouped_yes[feature_name].value_counts()
        
        feature_df.plot(kind='bar')
        plt.title("Feature Histogram - "+feature_name)
        plt.ylabel("Count")
        plt.xlabel("Feature unique values")
        plt.tight_layout()
```

```python
com = UnderstandingData(df)

com.plot_histogram_continuous("age",50)
com.plot_histogram_categorical("job")
com.plot_histogram_categorical("marital")
com.plot_histogram_categorical("education")
com.plot_histogram_categorical("default")
com.plot_histogram_categorical("housing")
com.plot_histogram_categorical("loan")
com.plot_histogram_categorical("contact")
com.plot_histogram_categorical("month")
com.plot_histogram_categorical("day_of_week")
com.plot_histogram_continuous("campaign", 30)
com.plot_histogram_continuous("pdays", 30)
com.plot_histogram_categorical("previous")
com.plot_histogram_categorical("poutcome")
com.plot_histogram_continuous("emp_var_rate", 50)
com.plot_histogram_continuous("cons_price_idx", 50)
com.plot_histogram_continuous("cons_conf_idx", 50)
com.plot_histogram_continuous("euribor3m", 50)
com.plot_histogram_continuous("nr_employed", 50)
```
